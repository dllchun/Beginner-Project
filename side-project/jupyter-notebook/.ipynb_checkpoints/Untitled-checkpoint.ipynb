```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.inlabonline.net/categories/%E7%A7%8B%E5%86%AC%E6%96%B0%E6%AC%BE"
response = requests.request('GET', url)
```

```python
html = response.text
soup = BeautifulSoup(html, 'html.parser')
product_items = soup.select('li.product-item')

master_list = []
for idx, product_item in enumerate(product_items):
    title = product_item.find('div', {'class': 'title'}).text.strip()
    price = product_item.find('div', {'class': 'price'}).text.strip()
    # item_dict = {}
    # item_dict['title'] = title[idx]
    # item_dict['price'] = price[idx]
    print(title)

master_list
```
